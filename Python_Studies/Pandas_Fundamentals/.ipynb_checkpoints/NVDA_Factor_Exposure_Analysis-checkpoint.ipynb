```python
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
import pandas_datareader.data as web
import statsmodels.api as sm

Ticker = 'NVDA'
start_date = '2019-01-01'
end_date = '2025-11-20'

print(f"--- Factor Exposure Analysis for NVDA ---")

stock_data = yf.download(Ticker, start=start_date, end=end_date, progress=False, auto_adjust=True)['Close']

if isinstance(stock_data, pd.Series):
    stock_data = stock_data.to_frame()

stock_returns = stock_data.resample('ME').last().pct_change().dropna()

if getattr(stock_returns.index, 'tz', None) is not None:
    stock_returns.index = stock_returns.index.tz_localize(None)
stock_returns.index = stock_returns.index.to_period('M')

ff_data = web.DataReader('F-F_Research_Data_Factors', 'famafrench', start=start_date, end=end_date)[0]
ff_data = ff_data / 100

if not isinstance(ff_data.index, pd.PeriodIndex):
    ff_data.index = ff_data.index.to_period('M')

if isinstance(stock_returns.columns, pd.MultiIndex):
    stock_returns.columns = stock_returns.columns.get_level_values(0)

data = pd.merge(stock_returns, ff_data, left_index=True, right_index=True)

data = data.rename(columns={data.columns[0]: 'Stock_Returns'})

print(f"Data successfully aligned. Analyzing {len(data)} months of data.")
data['Excess_Returns'] = data["Stock_Returns"] - data['RF']

X = data[['Mkt-RF', 'SMB', 'HML']]
X = sm.add_constant(X)
Y = data['Excess_Returns']

model = sm.OLS(Y, X).fit()
print(model.summary())
```

```python
##Alpha: NVDA generated significant monthly Alpha of 3.5% (p < 0.05), indicating outperformance unrelated to standard risk factors.

##Market Beta: A beta of ~1.78 confirms NVDA acts as a high-leverage play on the S&P 500.
```

```python
#Visualization of Data
```

```python
WINDOW = 12

rolling_covariance = data['Stock_Returns'].rolling(window=WINDOW).cov(data['Mkt-RF'])
rolling_variance = data['Mkt-RF'].rolling(window=WINDOW).var()
data['Rolling_Beta'] = rolling_covariance / rolling_variance

plt.figure(figsize=(12, 6))
plt.plot(data.index.to_timestamp(), data['Rolling_Beta'], color='#76b900', linewidth=2, label='NVDA Beta (12m)')

plt.axhline(1.0, color='gray', linestyle='--', linewidth=1, label='Market Average (1.0)')
plt.axhline(data['Rolling_Beta'].mean(), color='red', linestyle=':', linewidth=1, label='NVDA Average')

plt.title(f"Regime Change: NVDA Rolling {WINDOW}-Month Beta", fontsize=14, fontweight='bold')
plt.ylabel("Beta (Risk relative to S&P 500)")
plt.legend(loc='upper left')
plt.grid(True, alpha=0.3)

plt.show()
```
