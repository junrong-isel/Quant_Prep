```python
import numpy as np

#Let's say matrix A is A(m rows times n columns) and matrix B is B (n rows x p columns)
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

C = A @ B
print(C)
```

```python
#2nd Method: Columns
#3rd Method: Rows
```

```python
# "Columns of C are the combinations of columns of A"
print(f"{A @ B[:, 0]} is the 1st column of C")

# "Rows of C are the combinations of rows of B"
print(f"{A[0, :] @ B} is the 1st row of C")
```

```python
#4th Method: AB = sum of (cols of A) x (rows of B)
```

```python
# Outer product of Col 1 ofA and Row 1 of B
sum1 = np.outer(A[:, 0], B[0, :])

# Outer product of Col 2 ofA and Row 2 of B
sum2 = np.outer(A[:,1], B[1,:])

print(sum1 + sum2)
```
